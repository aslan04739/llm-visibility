```python
from __future__ import annotations

import argparse
import os
import textwrap
import time
import warnings
from math import pi
from pathlib import Path
from typing import Optional

import matplotlib.pyplot as plt
import pandas as pd
import requests
import seaborn as sns
from openai import OpenAI

try:
    from google import genai as google_genai
except Exception:
    google_genai = None

try:
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", FutureWarning)
        import google.generativeai as legacy_genai
except Exception:
    legacy_genai = None


API_TIMEOUT_SECONDS = 60
DEFAULT_GEMINI_MODEL = "gemini-2.5-flash"
DEFAULT_OPENAI_MODEL = "gpt-4o"
DEFAULT_JUDGE_MODEL = "gpt-4o-mini"
DEFAULT_PERPLEXITY_MODEL = "sonar"

PROMPTS = {
    "Bench - reputation": "Selon les avis en ligne, compare Riverly avec Le Boat, Samboat, Crisboat, Canalous, etc.",
    "Bench - boats": "Compare les bateaux de Riverly avec Le Boat, Samboat, Crisboat, Canalous.",
    "Bench - destinations": "Compare les destinations de Riverly avec Le Boat, Samboat, Crisboat, Canalous.",
    "Bench - prices": "Compare les prix de Riverly avec Le Boat, Samboat, Crisboat, Canalous.",
    "Bench - accessibility": "Compare l'accessibilité de Riverly avec Le Boat, Samboat, Crisboat, Canalous.",
    "Bench - reputation_global": "Compare la réputation globale de Riverly avec Le Boat, Samboat, Crisboat, Canalous.",
    "Bench - experience": "Compare l'expérience client de Riverly avec Le Boat, Samboat, Crisboat, Canalous.",
    "Brand - boats": "Quels sont les bateaux de Riverly ?",
    "Brand - destinations": "Quelles sont les destinations de Riverly ?",
    "Brand - prices": "Quels sont les prix de Riverly ?",
    "Brand - accessibility": "Riverly est-il accessible sans permis ?",
    "Brand - reputation": "Quelle est la réputation de Riverly ?",
    "Brand - experience": "Quelle est l'expérience client chez Riverly ?",
    "Brand - sustainability": "Riverly et ses engagements durables ?",
    "Brand - history": "Histoire et origines de Riverly ?",
}

DISPLAY_LABELS = {
    "Bench - reputation": "Comparaison de réputation",
    "Bench - boats": "Qualité et gamme des bateaux",
    "Bench - destinations": "Couverture des destinations",
    "Bench - prices": "Prix et compétitivité",
    "Bench - accessibility": "Accessibilité sans permis",
    "Bench - reputation_global": "Réputation globale",
    "Bench - experience": "Expérience client",
    "Brand - boats": "Offre Riverly : bateaux",
    "Brand - destinations": "Offre Riverly : destinations",
    "Brand - prices": "Prix Riverly",
    "Brand - accessibility": "Accessibilité Riverly",
    "Brand - reputation": "Réputation Riverly",
    "Brand - experience": "Expérience Riverly",
    "Brand - sustainability": "Engagements durables",
    "Brand - history": "Histoire et origine de Riverly",
}


class GeminiGateway:
    def __init__(self, api_key: str, model_name: str) -> None:
        self.model_name = model_name
        self.kind = None
        self.client = None

        if not api_key:
            return

        if google_genai is not None:
            self.kind = "modern"
            self.client = google_genai.Client(api_key=api_key)
            return

        if legacy_genai is not None:
            legacy_genai.configure(api_key=api_key)
            self.kind = "legacy"
            self.client = legacy_genai.GenerativeModel(model_name)

    def available(self) -> bool:
        return self.client is not None

    def generate(self, prompt: str) -> str:
        if not self.available():
            return "Erreur Gemini : clé API manquante"

        try:
            if self.kind == "modern":
                response = self.client.models.generate_content(
                    model=self.model_name,
                    contents=prompt,
                )
                return getattr(response, "text", "") or ""

            response = self.client.generate_content(prompt)
            return getattr(response, "text", "") or ""
        except Exception as exc:
            return f"Erreur Gemini : {exc}"


class AuditRunner:
    def __init__(
        self,
        target_brand: str,
        openai_key: Optional[str],
        gemini_key: Optional[str],
        perplexity_key: Optional[str],
        output_dir: Path,
        openai_model: str = DEFAULT_OPENAI_MODEL,
        gemini_model: str = DEFAULT_GEMINI_MODEL,
        judge_model: str = DEFAULT_JUDGE_MODEL,
        perplexity_model: str = DEFAULT_PERPLEXITY_MODEL,
    ) -> None:
        self.target_brand = target_brand
        self.output_dir = output_dir
        self.openai_model = openai_model
        self.judge_model = judge_model
        self.perplexity_model = perplexity_model
        self.perplexity_key = perplexity_key or ""

        self.client_openai = OpenAI(api_key=openai_key) if openai_key else None
        self.client_gemini = GeminiGateway(gemini_key or "", gemini_model)

        self.engines = {
            "ChatGPT (GPT-4o)": self.query_openai,
            "Gemini (Flash)": self.query_gemini,
            "Perplexity (RAG)": self.query_perplexity,
        }

    def query_openai(self, prompt: str) -> str:
        if self.client_openai is None:
            return "Erreur OpenAI : clé API manquante"

        try:
            response = self.client_openai.chat.completions.create(
                model=self.openai_model,
                messages=[{"role": "user", "content": prompt}],
                temperature=0.2,
                timeout=API_TIMEOUT_SECONDS,
            )
            return response.choices[0].message.content or ""
        except Exception as exc:
            return f"Erreur OpenAI : {exc}"

    def query_gemini(self, prompt: str) -> str:
        return self.client_gemini.generate(prompt)

    def query_perplexity(self, prompt: str, retries: int = 3) -> str:
        if not self.perplexity_key:
            return "Erreur Perplexity : clé API manquante"

        url = "https://api.perplexity.ai/chat/completions"
        headers = {
            "Authorization": f"Bearer {self.perplexity_key}",
            "Content-Type": "application/json",
        }
        payload = {
            "model": self.perplexity_model,
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.2,
        }

        for attempt_index in range(retries):
            try:
                response = requests.post(
                    url,
                    json=payload,
                    headers=headers,
                    timeout=API_TIMEOUT_SECONDS,
                )
                if response.status_code == 200:
                    return response.json()["choices"][0]["message"]["content"]
                if response.status_code == 429:
                    time.sleep(2 ** attempt_index)
                    continue
                return f"Erreur Perplexity HTTP {response.status_code}"
            except Exception:
                time.sleep(2)

        return "Timeout Perplexity"

    def analyze_response(self, text_response: str) -> tuple[float, str]:
        if self.client_openai is None:
            return 0.0, "Analyse impossible"

        if "Erreur" in text_response or "Timeout" in text_response:
            return 0.0, "Erreur de connexion API"

        eval_prompt = f"""
Analyse cette réponse d'une IA concernant la marque {self.target_brand}.
1. Donne une note globale de perception entre -1.0 (tres negatif / gros point faible) et +1.0 (tres positif / grand avantage). 0 = neutre.
2. Redige un resume direct de l'information principale en EXACTEMENT 10 a 15 mots maximum, en francais.
Reponse brute : "{text_response[:1500]}"
Format strict attendu : [Score numerique]|[Resume court]
"""

        try:
            response = self.client_openai.chat.completions.create(
                model=self.judge_model,
                messages=[{"role": "user", "content": eval_prompt}],
                temperature=0,
                timeout=15,
            )
            parts = (response.choices[0].message.content or "").strip().split("|")
            return float(parts[0]), parts[1].strip()
        except Exception:
            return 0.0, "Analyse impossible"

    def run_granular_audit(self) -> pd.DataFrame:
        print(
            f"Demarrage de l'audit 360 ({len(PROMPTS)} thematiques x 3 IA = {len(PROMPTS) * 3} requetes)..."
        )
        records = []

        for theme_name, prompt in PROMPTS.items():
            print(f"  Analyse en cours : {theme_name}")
            for engine_name, query_function in self.engines.items():
                raw_answer = query_function(prompt)
                score, summary = self.analyze_response(raw_answer)
                records.append(
                    {
                        "Axe Analysé": theme_name,
                        "Axe Lisible": DISPLAY_LABELS.get(theme_name, theme_name),
                        "Moteur IA": engine_name,
                        "Prompt Exact": prompt,
                        "Score Perception": score,
                        "Synthèse": summary,
                    }
                )
                time.sleep(1)

        return pd.DataFrame(records)

    def _save_current_figure(self, path: Path) -> None:
        plt.tight_layout()
        plt.savefig(path, dpi=300, bbox_inches="tight")
        plt.close()

    def generate_visuals(self, dataframe: pd.DataFrame) -> None:
        sns.set_theme(style="whitegrid", context="talk")
        dataframe = dataframe.copy()

        theme_order = (
            dataframe.groupby("Axe Lisible")["Score Perception"]
            .mean()
            .sort_values(ascending=True)
            .index.tolist()
        )

        # 1. Perception by theme and engine
        figure, axis = plt.subplots(figsize=(15, 10))
        sns.barplot(
            data=dataframe,
            y="Axe Lisible",
            x="Score Perception",
            hue="Moteur IA",
            order=theme_order,
            palette="Set2",
            ax=axis,
        )
        axis.axvline(0, color="black", linewidth=1.5, linestyle="--")
        axis.set_title(
            f"Perception & Competitivite de {self.target_brand} par thematique",
            fontweight="bold",
            pad=20,
        )
        axis.set_xlabel("Score (Negatif = Faiblesse | Positif = Force)")
        axis.set_ylabel("Theme analyse")
        axis.legend(title="Moteur IA", loc="lower right")
        self._save_current_figure(self.output_dir / "01_perception_scores_riverly.png")

        # 2. Verbatim matrix
        pivot_frame = dataframe.pivot(index="Axe Lisible", columns="Moteur IA", values="Synthèse").reindex(
            theme_order
        )
        figure, axis = plt.subplots(figsize=(22, 10))
        axis.axis("off")
        axis.axis("tight")

        row_labels = [textwrap.fill(label, 28) for label in pivot_frame.index]
        column_labels = pivot_frame.columns.tolist()
        cell_text = [[textwrap.fill(str(value), 42) for value in row] for row in pivot_frame.values]

        table = axis.table(
            cellText=cell_text,
            rowLabels=row_labels,
            colLabels=column_labels,
            cellLoc="left",
            loc="center",
            bbox=[0.15, 0, 0.85, 1],
        )
        table.auto_set_font_size(False)
        table.set_fontsize(9)
        table.scale(1, 3)

        for (row_index, col_index), cell in table.get_celld().items():
            if row_index == 0 or col_index == -1:
                cell.set_text_props(weight="bold", color="white")
                cell.set_facecolor("#2c3e50")
            else:
                cell.set_facecolor("#ecf0f1")

        plt.title(
            "Matrice des Insights : Benchmark vs Sentiment de Marque",
            fontweight="bold",
            size=16,
            pad=20,
        )
        plt.savefig(self.output_dir / "02_matrice_verbatims_riverly.png", dpi=300, bbox_inches="tight")
        plt.close()

        # 3. Consensus view, same format as agreed
        plt.figure(figsize=(14, 8))
        sns.stripplot(
            data=dataframe,
            x="Score Perception",
            y="Axe Lisible",
            hue="Moteur IA",
            order=theme_order,
            size=10,
            jitter=True,
            palette="Set1",
            alpha=0.8,
        )
        plt.title("Indice de Consensus des IA sur la marque et le marche", fontweight="bold", pad=20)
        plt.axvline(0, color="black", linewidth=1)
        plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
        plt.tight_layout()
        plt.savefig(self.output_dir / "03_consensus_ia_riverly.png", dpi=300, bbox_inches="tight")
        plt.close()

        # 4. Radar with scores from the three engines
        family_scores = dataframe.groupby(["Axe Analysé", "Moteur IA"])["Score Perception"].mean().reset_index()
        family_order = ["Bench", "Brand"]
        available_families = [family for family in family_order if family in family_scores["Axe Analysé"].str.split(" - ").str[0].unique()]

        if available_families:
            engine_names = ["ChatGPT (GPT-4o)", "Gemini (Flash)", "Perplexity (RAG)"]
            engine_palette = sns.color_palette("Set2", n_colors=len(engine_names))
            angles = [index / float(len(available_families)) * 2 * pi for index in range(len(available_families))]
            angles += angles[:1]

            figure, axis = plt.subplots(figsize=(12, 12), subplot_kw=dict(polar=True))
            axis.set_theta_offset(pi / 2)
            axis.set_theta_direction(-1)
            axis.set_rlabel_position(0)

            family_grouped = dataframe.assign(Famille=dataframe["Axe Analysé"].str.split(" - ").str[0]).groupby(
                ["Famille", "Moteur IA"]
            )["Score Perception"].mean().reset_index()

            for color, engine_name in zip(engine_palette, engine_names):
                engine_slice = family_grouped[family_grouped["Moteur IA"] == engine_name].set_index("Famille")[
                    "Score Perception"
                ]
                values = [float(engine_slice.get(family, 0)) for family in available_families]
                values += values[:1]
                axis.plot(angles, values, linewidth=2, label=engine_name, color=color)
                axis.fill(angles, values, alpha=0.10, color=color)

            axis.set_xticks(angles[:-1])
            axis.set_xticklabels(["Benchmark", "Marque"], size=10, fontweight="bold")
            axis.set_yticks([-1, -0.5, 0, 0.5, 1])
            axis.set_yticklabels(["Alerte", "Faible", "Neutre", "Bon", "Excellent"], color="grey", size=8)
            axis.set_ylim(-1, 1)
            axis.set_title(
                f"Radar 360 : Benchmark & ADN {self.target_brand}",
                size=16,
                fontweight="bold",
                pad=40,
            )
            axis.legend(loc="upper right", bbox_to_anchor=(1.25, 1.1))
            self._save_current_figure(self.output_dir / "04_radar_360_riverly.png")

        dataframe.to_csv(self.output_dir / "audit_360_riverly.csv", index=False)
        print("\n[SUCCES] Les 4 graphiques ont ete generes avec succes.")


def build_argument_parser() -> argparse.ArgumentParser:
    parser = argparse.ArgumentParser(description="Audit 360 Riverly: prompting, scoring and executive visuals.")
    parser.add_argument("--target-brand", default=os.getenv("TARGET_BRAND", "Riverly"), help="Brand to audit")
    parser.add_argument("--openai-key", default=os.getenv("OPENAI_API_KEY", ""), help="OpenAI API key")
    parser.add_argument("--gemini-key", default=os.getenv("GEMINI_API_KEY", ""), help="Gemini API key")
    parser.add_argument("--perplexity-key", default=os.getenv("PERPLEXITY_API_KEY", ""), help="Perplexity API key")
    parser.add_argument(
        "--output-dir",
        default=os.getcwd(),
        help="Directory where the CSV and PNG files will be written",
    )
    parser.add_argument("--openai-model", default=DEFAULT_OPENAI_MODEL, help="OpenAI completion model")
    parser.add_argument("--gemini-model", default=DEFAULT_GEMINI_MODEL, help="Gemini model")
    parser.add_argument("--judge-model", default=DEFAULT_JUDGE_MODEL, help="OpenAI judge model")
    parser.add_argument("--perplexity-model", default=DEFAULT_PERPLEXITY_MODEL, help="Perplexity model")
    return parser


def main() -> None:
    arguments = build_argument_parser().parse_args()
    output_dir = Path(arguments.output_dir).resolve()
    output_dir.mkdir(parents=True, exist_ok=True)

    missing_keys = []
    if not arguments.openai_key:
        missing_keys.append("OPENAI_API_KEY")
    if not arguments.gemini_key:
        missing_keys.append("GEMINI_API_KEY")
    if not arguments.perplexity_key:
        missing_keys.append("PERPLEXITY_API_KEY")

    if missing_keys:
        print("Cle keys manquantes : " + ", ".join(missing_keys))
        print("Le script peut quand meme tourner si tu passes les cles en variables d'environnement ou en argument.")
        return

    runner = AuditRunner(
        target_brand=arguments.target_brand,
        openai_key=arguments.openai_key,
        gemini_key=arguments.gemini_key,
        perplexity_key=arguments.perplexity_key,
        output_dir=output_dir,
        openai_model=arguments.openai_model,
        gemini_model=arguments.gemini_model,
        judge_model=arguments.judge_model,
        perplexity_model=arguments.perplexity_model,
    )
    results_frame = runner.run_granular_audit()
    runner.generate_visuals(results_frame)


if __name__ == "__main__":
    main()
```
